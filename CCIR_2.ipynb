## Import libraries and load data

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import wilcoxon, pearsonr, spearmanr, rankdata, linregress
```

```python
times_df = pd.read_csv('CCIR-laptimes.csv')

# Convert time string to total seconds 
def time_to_seconds(time_str): 
    minute, sec_millisec = time_str.split(':') 
    seconds, milliseconds = sec_millisec.split('.') 
    total_seconds = int(minute) * 60 + int(seconds) + float(milliseconds) / 1000 
    return total_seconds 
times_df['Lap-Time'] = times_df['Time'].apply(time_to_seconds)

times_df
```

```python
bios_df = pd.read_csv('CCIR-biometrics.csv')
bios_df.dropna(subset=['Intervention'], inplace=True)
bios_df
```

```python
# Define a color palette with one color per participant
palette = sns.color_palette("husl", len(times_df['ID'].unique()))
```

## Lap-times

### Fastest lap-time by intervention

```python
# Group by ID and Intervention to get the fastest lap time
fastest_laps = times_df.groupby(['ID', 'Intervention'])['Lap-Time'].min().reset_index()

# Use color palette 
participant_colors = {participant_id: palette[i] for i, participant_id in enumerate(fastest_laps['ID'].unique())}

# Create the plot
sns.set(style="whitegrid")
plt.figure(figsize=(6, 4))

# Plot lines for individual participants
for participant_id in fastest_laps['ID'].unique():
    participant_data = fastest_laps[fastest_laps['ID'] == participant_id]
    sns.lineplot(x='Intervention', y='Lap-Time', data=participant_data, marker='o', linestyle='-', color=participant_colors[participant_id], label=f'Participant {participant_id}', alpha=0.6)

# Plot mean and standard deviation of the fastest laps with error bars
mean_data = fastest_laps.groupby('Intervention')['Lap-Time'].agg(['mean', 'std']).reset_index()
sns.lineplot(x='Intervention', y='mean', data=mean_data, marker='o', color='black', linewidth=2, label='Mean ± SD', zorder=10)
plt.errorbar(mean_data['Intervention'], mean_data['mean'], yerr=mean_data['std'], fmt='o', color='black', capsize=5, elinewidth=2, zorder=5)

plt.title('Fastest Lap Time with vs without Intervention', fontsize=12)
plt.xlabel('Intervention', fontsize=12)
plt.ylabel('Fastest Lap Time (s)', fontsize=12)
plt.yticks(fontsize=10)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)
plt.tight_layout()
plt.show()
```

##### Wilcoxon Signed Rank Test

```python
# Group by ID and Intervention to get the fastest lap time
fastest_laps = times_df.groupby(['ID', 'Intervention'])['Lap-Time'].min().reset_index()

# Extract the lap times for each participant under intervention and no intervention
intervention_times = fastest_laps[fastest_laps['Intervention'] == 'Yes']['Lap-Time'].values
no_intervention_times = fastest_laps[fastest_laps['Intervention'] == 'No']['Lap-Time'].values

# Run the Wilcoxon signed-rank test
stat, p = wilcoxon(intervention_times, no_intervention_times)
print(f'Wilcoxon test statistic: {stat}, p-value: {p}')

# Calculate median lap times for intervention and no intervention
median_no_intervention = np.median(no_intervention_times)
median_intervention = np.median(intervention_times)
print(f'Median fastest lap time with intervention: {round(median_intervention, 2)} seconds')
print(f'Median fastest lap time without intervention: {round(median_no_intervention, 2)} seconds')

# Calculate MEAN (instead of median) lap times for intervention and no intervention
mean_no_intervention = np.mean(no_intervention_times)
mean_intervention = np.mean(intervention_times)
print(f'Mean lap time with intervention: {round(mean_intervention, 2)} seconds')
print(f'Mean lap time without intervention: {round(mean_no_intervention, 2)} seconds')
```

```python
print("\nResults of the Wilcoxon Signed-Rank Test\n")
print(f"Median lap time was {round(median_intervention, 2)} seconds with intervention and {round(median_no_intervention, 2)} seconds without intervention\n"
      f"A Wilcoxon signed-rank test was conducted to compare the lap times with and without intervention.\n"
      f"The results indicated that the difference in lap times was not statistically significant, W = {stat}, p = {round(p, 3)}.\n"
      f"This suggests that the visualization intervention did not have a significant effect on the participants' racing track times.")
```

```python
plt.boxplot([no_intervention_times, intervention_times], labels=['No Intervention', 'Intervention'])
plt.scatter([1], [mean_no_intervention], color='red', marker='o', label='Mean - No Intervention')
plt.scatter([2], [mean_intervention], color='red', marker='o', label='Mean - Intervention')
plt.title('Comparison of Fastest Lap Times with and without Intervention')
plt.ylabel('Mean Fastest Lap Time (seconds)')
plt.show()
```

### Fastest lap-time by stress resilience

```python
# Group by ID and Intervention to get the fastest lap time
fastest_laps = times_df.groupby(['ID', 'Intervention'])['Lap-Time'].min().reset_index()

# Merge the fastest lap times with the bios_df on ID
merged_df = pd.merge(fastest_laps, bios_df[['ID', 'Stress resilience']], on='ID')

# Define a function to calculate and return Spearman's r and p-value
def get_spearman_stats(x, y):
    spearman_r, p_value = spearmanr(x, y)
    return f'Spearman r={spearman_r:.2f}, p={p_value:.3f}'

# Create the plot
sns.set(style="whitegrid")
plt.figure(figsize=(6, 5))

# Define a color palette for the Intervention
intervention_palette = {'Yes': 'blue', 'No': 'green'}

# Plot a scatter plot of fastest lap time as a function of stress resilience
# Different colors and markers for Intervention and No Intervention
sns.scatterplot(x='Stress resilience', y='Lap-Time', data=merged_df, hue='Intervention', palette=intervention_palette, s=100, alpha=0.7)

# Calculate and store Spearman's r and p-value for each group
stats_yes = get_spearman_stats(merged_df[merged_df['Intervention'] == 'Yes']['Stress resilience'],
                               merged_df[merged_df['Intervention'] == 'Yes']['Lap-Time'])

stats_no = get_spearman_stats(merged_df[merged_df['Intervention'] == 'No']['Stress resilience'],
                              merged_df[merged_df['Intervention'] == 'No']['Lap-Time'])

# Plot regression lines for Intervention and No Intervention separately
sns.regplot(x='Stress resilience', y='Lap-Time', data=merged_df[merged_df['Intervention'] == 'Yes'], scatter=False, color='blue', ci=None)
sns.regplot(x='Stress resilience', y='Lap-Time', data=merged_df[merged_df['Intervention'] == 'No'], scatter=False, color='green', ci=None)

# Create custom legend
handles, labels = plt.gca().get_legend_handles_labels()
custom_labels = [f'No Intervention: {stats_no}', f'Intervention (Yes): {stats_yes}']
plt.legend(handles, custom_labels, fontsize=10)

plt.title('Fastest Lap Time as a Function of Stress Resilience', fontsize=14)
plt.xlabel('Stress Resilience', fontsize=12)
plt.ylabel('Fastest Lap Time (s)', fontsize=12)
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
